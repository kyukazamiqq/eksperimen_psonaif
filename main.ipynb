```python
#import modul cihuy
import pandas as pd 
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Load dan preprocessing data
data = load_breast_cancer()
data = pd.DataFrame(data.data, columns=data.feature_names)
```

```python
data.head()
```

```python
data["target"] = load_breast_cancer().target
data["target_names"] = load_breast_cancer().target_names[data.target]
```

```python
#Split fitur ke variabel x dan y, x itu fitur y itu target
X = data.drop(columns=["target", "target_names"])
y = data["target"]
#Normalisasi Sederhana aj
X = StandardScaler().fit_transform(X)
```

```python
# Parameter PSO
n_particles = 30
n_iterations = 100
n_features = X.shape[1]
w = 0.72  # Inertia weight
c1, c2 = 1.5, 1.5  # Cognitive & social components

# Inisialisasi partikel
positions = np.random.randint(0, 2, (n_particles, n_features))
velocities = np.random.rand(n_particles, n_features)
personal_best_positions = positions.copy()
personal_best_scores = np.zeros(n_particles)
```

```python
# Inisialisasi partikel
positions = np.random.randint(0, 2, (n_particles, n_features))
velocities = np.random.rand(n_particles, n_features)
personal_best_positions = positions.copy()
personal_best_scores = np.zeros(n_particles)

# Fungsi fitness: akurasi Naive Bayes
def fitness(position):
    if np.count_nonzero(position) == 0:
        return 0
    subset = X[:, position.astype(bool)]
    clf = GaussianNB()
    scores = cross_val_score(clf, subset, y, cv=5)
    return scores.mean()

# Hitung personal best awal
for i in range(n_particles):
    personal_best_scores[i] = fitness(positions[i])

# Global best
global_best_idx = np.argmax(personal_best_scores)
global_best_position = personal_best_positions[global_best_idx].copy()
```

```python
# PSO Main loop
for iteration in range(n_iterations):
    for i in range(n_particles):
        r1, r2 = np.random.rand(n_features), np.random.rand(n_features)
        velocities[i] = (w * velocities[i] +
                         c1 * r1 * (personal_best_positions[i] - positions[i]) +
                         c2 * r2 * (global_best_position - positions[i]))
        
        sigmoid = 1 / (1 + np.exp(-velocities[i]))
        positions[i] = np.where(np.random.rand(n_features) < sigmoid, 1, 0)

        score = fitness(positions[i])
        if score > personal_best_scores[i]:
            personal_best_scores[i] = score
            personal_best_positions[i] = positions[i].copy()
    
    global_best_idx = np.argmax(personal_best_scores)
    global_best_position = personal_best_positions[global_best_idx].copy()

    print(f"Iterasi {iteration+1} - Akurasi Terbaik: {personal_best_scores[global_best_idx]:.4f}")
```

```python
# Final model training
final_features = global_best_position.astype(bool)
print("\nFitur yang dipilih:", np.where(final_features)[0])
```

```python
selected_features = [1, 2, 4, 9, 11, 18, 21, 22, 23, 27,28]

X_selected = data.iloc[:, selected_features]
```

```python
#Train Modelnya uhuy
# Bagi data jadi train & test
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Latih model Naive Bayes
model = GaussianNB()
model.fit(X_train, y_train)

# Prediksi dan evaluasi
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f"Akurasi model Naive Bayes (fitur terpilih): {accuracy:.4f}")
```
